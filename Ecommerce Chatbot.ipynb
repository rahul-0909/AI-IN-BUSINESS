```python
import re
import random
from difflib import get_close_matches
from random import randrange

intents={
  "intents": [
    {
      "intent": "greeting",
      "patterns": ["Hi", "Hello", "Hey there"],
      "responses": ["Hello! How can I help you today?", "Hi there! What can I do for you?", "Hey, how can I assist you?"]
    },
    {
      "intent": "goodbye",
      "patterns": ["Bye", "Goodbye", "See you later"],
      "responses": ["Goodbye! Have a great day.", "See you later. Take care!", "Bye for now!"]
    },
    {
      "intent": "thanks",
      "patterns": ["Thank you", "Thanks a lot", "Appreciate it"],
      "responses": ["You're welcome! Is there anything else I can help you with?", "Glad to be of assistance. Let me know if you need anything else.", "No problem. Don't hesitate to reach out if you need anything else."]
    },
    {
      "intent": "product_search",
      "patterns": ["I want to buy a [product_name]", "Can you help me find a [product_name]?", "Do you have any [product_name]?","search [product_name]"],
      "responses": ["Sure, let me check if we have any such product.", "Let me see if we have any such product in stock."]
    },
    {
      "intent": "support",
      "patterns": ["I need help with my order", "Can you assist me with an issue?", "I have a question about my account"],
      "responses": ["I'm here to help. What do you need assistance with?", "What can I help you with?", "I'm happy to help. What's the issue you're experiencing?"]
    },
    {
      "intent": "feedback",
      "patterns": ["I have some feedback for you", "I want to leave feedback", "Can I provide feedback?"],
      "responses": ["Thank you for taking the time to share your feedback with us. We value your input.", "We appreciate your feedback and will use it to improve our services.", "Sure, we welcome your feedback. What would you like to share with us?"]
    },
    {
      "intent": "OrderTracking",
      "patterns": ["Where is my order?","When will my package be delivered?","What about my order?"],
      "responses": ["Your order has been shipped and is on its way to your address. You can track it using the tracking number provided in the confirmation email.",
        "Your package is expected to be delivered by the end of the day. You can check the delivery status using the tracking number provided in the confirmation email."
   ]
    },
    {
      "intent": "SalesAndPromotions",
      "patterns": ["Are there any discounts available?","Can you give me a coupon code for my purchase?","Is there any discount available?"],
      "responses": ["Yes, we are currently offering a 10% discount on all orders over $50. Use the code 'SAVE10' at checkout.",
        "Yes, here is a coupon code for 15% off your next purchase: 'SPRINGSALE15'."]
    },
    {
      "intent": "Payment",
      "patterns": ["How can I pay ?", "What payment methods do you accept?"],
      "responses": ["We accept various payment methods, including credit cards, debit cards, PayPal, and gift cards. Which payment method would you like to use?", "You can pay for your order using any of our accepted payment methods.We accept various payment methods, including credit cards, debit cards, PayPal, and gift cards."]
    },
    {
  "intent": "refunds",
  "patterns": ["How can I get a refund?", "What's your refund policy?", "Can I return this item for a refund?","I want my refund."],
  "responses": ["If you're not satisfied with your purchase, we offer a refund within 30 days of the purchase date. Would you like me to initiate a refund for you now?", "Our refund policy allows for returns within 30 days of purchase. You can initiate a refund by following the instructions on our website or by contacting customer support.", "You can return a item for a refund within 30 days of purchase. Please initiate the return process on our website or by contacting customer support."]
}
  ]
}


# define a function to tokenize user input
def tokenize(text):
    text = text.lower()  # convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    tokens = text.split()  # split into tokens
    return tokens

# define a function to compute similarity between two sentences
def similarity(sent1, sent2):
    tokens1 = set(tokenize(sent1))
    tokens2 = set(tokenize(sent2))
    score = len(tokens1.intersection(tokens2)) / len(tokens1.union(tokens2))
    return score

# define a function to find the best matching intent based on user input
def find_intent(text):
    max_score = 0
    best_match = None
    for intent in intents['intents']:
        for pattern in intent['patterns']:
            score = similarity(text, pattern)
            if score > max_score:
                max_score = score
                best_match = intent
    return best_match

# define a function to generate a response based on the detected intent
def generate_response(intent):
    responses = intent['responses']
    response = random.choice(responses)  # choose the random response by default
    if len(responses) > 1:
        response = responses[randrange(len(responses))]  # choose a random response if there are multiple options
    return response

# define a function to handle user input and generate a response
def handle_input(text):
    intent = find_intent(text)
    if intent is not None:
        response = generate_response(intent)
        return response
    else:
        return "Sorry, I didn't understand that. Can you please rephrase?"

# example usage
while True:
    user_input = input('You: ')
    response = handle_input(user_input)
    print('Bot:', response)                                         
    print()
```
